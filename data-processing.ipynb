Find the number of missing values in each column and each row. Remove rows where at least 50% of the values are missing. Then remove columns where at least 50% of the values are missing.

```python
for column in songs.columns:
    #print(column, songs[column].isna().sum(),len(songs)/2)
    num_null  = songs[column].isna().sum()

    if num_null > (len(songs)/2):
        songs = songs.drop(columns=[column])

songs.columns
```

Calculate descriptive statistics for each column. Let's define an outlier as a value at least 3 standard deviations away from the mean. Which columns have outliers? What are those values?

```python
songs.describe()
```

```python
def get_outliers(column):
    three_sd = 3*(songs[column].std())
    mean = songs[column].sum()/len(songs)
    max = mean + three_sd
    min = mean - three_sd
    vals = []
    for i in range(len(songs)):
        val = songs[column].iloc[i]
        if val > max:
            vals.append(val)
            #songs.at[i,column] = max
        elif val < min:
            vals.append(val)
            #songs.at[i,column] = min
    print(column, vals)
```

```python
for column in songs.describe().columns:
    get_outliers(column)
```

With the remaining columns, use scikit-learn to impute missing values. For continuous features, fill in the mean. For categorical features, fill in the mode.

```python
songs.info()
```

```python
imp_mean = SimpleImputer(missing_values= np.nan, strategy='mean')
num_cols=['Beats Per Minute (BPM)','Energy','Loudness (dB)','Liveness','Length (Duration)','Acousticness','Popularity','Speechiness']
for col in num_cols:
    songs[col]=imp_mean.fit_transform(songs[[col]])

songs.info()
```

```python
imp_mode = SimpleImputer(missing_values= np.nan, strategy='most_frequent')
cat_cols=['Title','Artist','Top Genre','Year','Month']
for col in cat_cols:
    new_col = imp_mode.fit_transform(songs[[col]])
    songs[col]=new_col
songs.info()
```

Combine the date-related columns into one column with the Pandas to_datetime() method, then use that column to create a numeric Age column (in years). Calculate Age based on today's date; it doesn't have to be a whole number. Once you've created the Age column, remove the other date-related columns, including the one you created with Pandas.

```python
songs['Day'] = 1
songs['Date']=pd.to_datetime(songs[['Year','Month','Day']])

songs['Today'] = pd.to_datetime('2022-9-6')

songs['Age']=songs['Today']-songs['Date']
songs.dtypes
```

```python
songs=songs.drop(columns=['Year','Month','Day','Date','Today'])
songs.head()
```

Create dummy variables for the categorical features. Drop one level of each feature to end up with k-1 dummies, not k.
